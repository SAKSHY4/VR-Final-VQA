```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from itertools import combinations

def calculate_diversity_metrics(df, categorical_columns):
    """
    Calculate diversity metrics for categorical variables
    """
    results = {}

    # 1. Individual column entropy (variation)
    for col in categorical_columns:
        value_counts = df[col].value_counts(normalize=True)
        entropy = -np.sum(value_counts * np.log2(value_counts))
        results[f'{col}_entropy'] = entropy / np.log2(len(value_counts))  # Normalized entropy

    # 2. Pairwise combination coverage
    for col1, col2 in combinations(categorical_columns, 2):
        # Count unique combinations
        combo_df = df.groupby([col1, col2]).size().reset_index(name='count')
        # Calculate coverage ratio (actual vs theoretical combinations)
        theoretical_combos = df[col1].nunique() * df[col2].nunique()
        actual_combos = len(combo_df)
        results[f'{col1}_{col2}_coverage'] = actual_combos / theoretical_combos

        # Calculate combination entropy
        combo_counts = combo_df['count'] / combo_df['count'].sum()
        combo_entropy = -np.sum(combo_counts * np.log2(combo_counts))
        results[f'{col1}_{col2}_entropy'] = combo_entropy / np.log2(len(combo_counts))

    return results
```

```python
def stratified_diversity_sample(df, categorical_columns, n_samples=10000):
    """
    Create a stratified sample that maximizes diversity
    """
    # 1. Cap dominant categories
    sample_frames = []

    # Handle product type diversity
    product_types = df['product_type'].value_counts()
    max_per_type = min(500, n_samples // product_types.shape[0])

    # Sample from each product type with a cap
    for p_type, count in product_types.items():
        # FIX: Use .isin() instead of == when comparing with a potential list
        if isinstance(p_type, list):
            type_df = df[df['product_type'].isin([p_type])]
        else:
            type_df = df[df['product_type'] == p_type]

        sample_size = min(count, max_per_type)

        # Further stratify by brand within each product type
        type_brands = type_df['brand'].value_counts()
        if len(type_brands) > 1:
            brand_samples = []
            max_per_brand = sample_size // min(len(type_brands), 5)

            for brand, b_count in type_brands.items():
                # FIX: Same check for brands
                if isinstance(brand, list):
                    brand_df = type_df[type_df['brand'].isin([brand])]
                else:
                    brand_df = type_df[type_df['brand'] == brand]

                brand_sample_size = min(b_count, max_per_brand)
                brand_samples.append(brand_df.sample(brand_sample_size, random_state=42))

            # Combine brand samples
            type_sample = pd.concat(brand_samples)

            # If we need more samples from this product type
            if len(type_sample) < sample_size:
                remaining = sample_size - len(type_sample)
                # Sample from remaining records not included in brand samples
                remaining_df = type_df[~type_df.index.isin(type_sample.index)]
                if len(remaining_df) > 0:
                    type_sample = pd.concat([type_sample, remaining_df.sample(min(remaining, len(remaining_df)), random_state=42)])
        else:
            # If only one brand, just sample directly
            type_sample = type_df.sample(sample_size, random_state=42)

        sample_frames.append(type_sample)

    # Combine all samples
    final_sample = pd.concat(sample_frames)

    # If we have more than requested, randomly subsample
    if len(final_sample) > n_samples:
        final_sample = final_sample.sample(n_samples, random_state=42)

    return final_sample
```

```python
def identify_optimal_combinations(df, categorical_columns, top_n=5):
    """
    Identify category combinations with highest diversity and representativeness

    Parameters:
    -----------
    df : pandas DataFrame
        The cleaned dataset
    categorical_columns : list
        List of cleaned column names to analyze
    top_n : int
        Number of top combinations to return

    Returns:
    --------
    list of tuples
        (column_combination, metrics_dict)
    """
    import numpy as np
    from itertools import combinations

    # Calculate metrics for all possible combinations
    all_metrics = {}

    # For each possible subset of columns (size 2-3)
    for k in range(2, min(4, len(categorical_columns)+1)):
        for cols in combinations(categorical_columns, k):
            # Create subset dataframe
            subset_df = df[list(cols)].copy()

            # 1. Calculate entropy of combinations
            # Convert combinations to strings for counting
            subset_df['combo'] = subset_df[list(cols)].astype(str).agg('-'.join, axis=1)
            subset_counts = subset_df['combo'].value_counts(normalize=True)
            entropy = -np.sum(subset_counts * np.log2(subset_counts + 1e-10))
            max_entropy = np.log2(len(subset_counts)) if len(subset_counts) > 1 else 1
            norm_entropy = entropy / max_entropy if max_entropy > 0 else 0

            # 2. Calculate coverage
            total_combos = 1
            for col in cols:
                total_combos *= df[col].nunique()
            actual_combos = len(subset_counts)
            coverage = actual_combos / total_combos if total_combos > 0 else 0

            # 3. Calculate balance score (how evenly distributed)
            balance = 1 - (subset_counts.std() / (subset_counts.mean() + 1e-10)) if len(subset_counts) > 1 else 0

            # Composite score (weighted average)
            composite_score = 0.4 * norm_entropy + 0.4 * coverage + 0.2 * balance

            all_metrics[cols] = {
                'entropy': norm_entropy,
                'coverage': coverage,
                'balance': balance,
                'composite_score': composite_score
            }

    # Sort by composite score and return top N
    sorted_combos = sorted(all_metrics.items(), key=lambda x: x[1]['composite_score'], reverse=True)
    return sorted_combos[:top_n]

def report_category_variation(df, categorical_columns):
    """
    Report individual category variation metrics
    """
    variation_metrics = {}

    for col in categorical_columns:
        counts = df[col].value_counts(normalize=True)

        # Calculate metrics
        unique_count = counts.shape[0]
        entropy = -np.sum(counts * np.log2(counts + 1e-10))
        norm_entropy = entropy / np.log2(unique_count)
        gini = 1 - np.sum(counts**2)
        top1_dominance = counts.iloc[0]
        top5_concentration = counts.iloc[:5].sum() if len(counts) >= 5 else counts.sum()

        variation_metrics[col] = {
            'unique_values': unique_count,
            'entropy': norm_entropy,
            'gini': gini,
            'top1_dominance': top1_dominance,
            'top5_concentration': top5_concentration
        }

    return pd.DataFrame(variation_metrics).T.sort_values('entropy', ascending=False)
```

```python
import pandas as pd
import numpy as np
from collections import defaultdict

def extract_text_value(item):
    """Extract clean values from nested structures, optimized for text fields"""
    if isinstance(item, list):
        if len(item) > 0:
            if isinstance(item[0], dict) and 'value' in item[0]:
                return item[0]['value']
            return item[0]
    elif isinstance(item, dict) and 'node_id' in item:
        # Special handling for node paths
        return item.get('path', str(item.get('node_id')))
    return item

def balanced_multi_stage_sampling(df, n_samples=12000):
    """
    Multi-stage sampling optimized for the ABO dataset with unique main_image_id guarantee
    """
    print("Starting balanced sampling process...")
    print(f"Input dataset size: {len(df)}")

    # Track sampled main_image_ids to avoid duplicates
    sampled_image_ids = set()

    # ---- PRE-PROCESSING STEP: Group by main_image_id ----
    # Create a lookup from main_image_id to item_ids
    print("Creating image-to-item mapping...")
    image_to_items = df.groupby('main_image_id')['item_id'].apply(list).to_dict()

    # For each image_id with multiple items, select one representative item
    # (can choose random, first, or best based on your criteria)
    unique_image_items = {}
    for img_id, item_list in image_to_items.items():
        if pd.isna(img_id):
            continue  # Skip missing image IDs
        # Choose the first item by default (you could modify this logic)
        unique_image_items[img_id] = item_list[0]

    # Create a deduplicated dataframe with one item per image
    clean_df = df.drop_duplicates(subset=['main_image_id'])

    print(f"After deduplication: {len(clean_df)} unique images available")
    if len(clean_df) < n_samples:
        print(f"WARNING: Only {len(clean_df)} unique images available, less than requested {n_samples}")

    # Unchanged - extract clean values for all columns
    primary_cols = ['item_keywords', 'node', 'bullet_point']
    secondary_cols = ['product_type', 'brand', 'color', 'material']
    all_cols = primary_cols + secondary_cols

    for col in all_cols:
        if col in clean_df.columns:
            clean_df[f'{col}_clean'] = clean_df[col].apply(extract_text_value)

    # Handle missing values (unchanged)
    for col in [f'{c}_clean' for c in all_cols if f'{c}_clean' in clean_df.columns]:
        clean_df[col] = clean_df[col].fillna('MISSING')

    # Updated caps with more specific restrictions
    caps = {
        'product_type_clean': {
            'CELLULAR_PHONE_CASE': 500,  # Reduced from 1200
            'SHOES': 700,               # Added specific cap for shoes
            'BOOT': 200,                # Added cap for boots
            'SANDAL': 200,              # Added cap for sandals
        },
        'brand_clean': {
            'Amazon Brand - Solimo': 700,    # Slightly increased
            'AmazonBasics': 1200,            # Reduced from 2667
            'find.': 500,                    # Added cap
            'Amazon Collection': 500,        # Added cap
            'Rivet': 500,                    # Added cap
            'Stone & Beam': 500,             # Added cap
        },
        'color_clean': {
            'Others': 200,              # Reduced
            'Multicolor': 150,          # Reduced
            'multi-colored': 100,       # Reduced
            'MISSING': 2500,            # Added cap for missing values
        }
    }

    # STAGE 1: Primary stratification (60%) based on optimal text attributes
    stage1_size = int(0.6 * n_samples)
    print(f"Stage 1: Sampling {stage1_size} items based on text attributes")

    # Create clusters based on text attributes
    try:
        # Create simple hash features from text fields
        for col in [f'{c}_clean' for c in primary_cols if f'{c}_clean' in clean_df.columns]:
            # Convert to string and take first 50 chars for hashing
            clean_df[f'{col}_hash'] = clean_df[col].astype(str).str[:50]

        # Create strata key based on available columns
        hash_cols = [f'{c}_clean_hash' for c in primary_cols if f'{c}_clean_hash' in clean_df.columns]
        if hash_cols:
            clean_df['text_strata'] = clean_df[hash_cols].apply(
                lambda x: hash(tuple(x)), axis=1
            ) % 1000  # Mod to create ~1000 strata

            # Sample from each stratum
            strata_counts = clean_df['text_strata'].value_counts()
            strata_roots = strata_counts.apply(np.sqrt)  # Power law to boost small strata
            strata_alloc = (strata_roots / strata_roots.sum() * stage1_size).astype(int)

            stage1_frames = []
            for stratum, size in strata_alloc.items():
                if size > 0:
                    stratum_df = clean_df[clean_df['text_strata'] == stratum]
                    # Apply caps within stratum
                    for col, cap_dict in caps.items():
                        for value, cap in cap_dict.items():
                            stratum_cap = max(1, int(cap * size / stage1_size))
                            over_count = (stratum_df[col] == value).sum()
                            if over_count > stratum_cap:
                                exclude = stratum_df[stratum_df[col] == value].sample(
                                    over_count - stratum_cap, random_state=42
                                ).index
                                stratum_df = stratum_df[~stratum_df.index.isin(exclude)]

                    # Sample from adjusted stratum
                    sample_size = min(size, len(stratum_df))
                    if sample_size > 0:
                        stage1_frames.append(stratum_df.sample(sample_size, random_state=42))

            stage1 = pd.concat(stage1_frames) if stage1_frames else pd.DataFrame()
        else:
            # Fallback if text columns aren't available
            stage1 = clean_df.sample(stage1_size, random_state=42)

    except Exception as e:
        print(f"Error in Stage 1: {e}")
        # Fallback to simpler approach
        stage1 = clean_df.sample(stage1_size, random_state=42)

    print(f"Stage 1 complete: {len(stage1)} items sampled")

    # After Stage 1 sampling, before Stage 2
    print("Prioritizing items with color information...")

    # Calculate how many items have missing color
    missing_color_count = sum(1 for i, row in stage1.iterrows() if row['color_clean'] == 'MISSING')
    missing_color_pct = missing_color_count / len(stage1) if len(stage1) > 0 else 0
    print(f"Stage 1 has {missing_color_pct:.1%} items with missing color")

    # If more than 30% have missing color, try to replace some
    if missing_color_pct > 0.3 and 'color_clean' in stage1.columns:
        # Identify items with missing color that could be replaced
        to_replace = min(int(missing_color_count * 0.3), 1000)  # Replace up to 30% of missing colors
        replace_candidates = stage1[stage1['color_clean'] == 'MISSING'].sample(to_replace, random_state=42)

        # Find items with color in remaining dataset
        remaining_with_color = remaining_df[remaining_df['color_clean'] != 'MISSING']

        # Replace if we have enough candidates
        if len(remaining_with_color) > 0:
            replacements = remaining_with_color.sample(min(len(remaining_with_color), len(replace_candidates)), random_state=42)

            # Remove the candidates from stage1
            stage1 = stage1[~stage1.index.isin(replace_candidates.index)]

            # Add the replacements
            stage1 = pd.concat([stage1, replacements])
            print(f"Replaced {len(replacements)} items missing color with items that have color information")

    # STAGE 2: Product type coverage (30%)
    stage2_size = int(0.3 * n_samples)
    print(f"Stage 2: Ensuring product type diversity with {stage2_size} items")

    # Skip already sampled items
    remaining_df = clean_df[~clean_df.index.isin(stage1.index)]

    # More balanced product type allocation
    # First, check what we already have from Stage 1
    stage1_product_counts = stage1['product_type_clean'].value_counts()

    # Get top 40 product types (increased from 30)
    product_types = clean_df['product_type_clean'].value_counts().head(40).index

    # Calculate target counts - lower for already overrepresented types
    target_counts = {}
    base_count = stage2_size // len(product_types)  # Base allocation

    for p_type in product_types:
        current_count = stage1_product_counts.get(p_type, 0)
        if p_type == 'SHOES':
            # Special handling for SHOES to hit our ~700 target
            target = max(0, 700 - current_count)
        elif p_type in caps.get('product_type_clean', {}):
            # Honor caps for specified types
            target = max(0, caps['product_type_clean'][p_type] - current_count)
        else:
            # Standard allocation for other types
            target = max(0, base_count - current_count // 2)  # Reduce allocation if already well-represented

        target_counts[p_type] = target

    # Normalize to ensure we add up to stage2_size
    total_target = sum(target_counts.values())
    if total_target > 0:  # Avoid division by zero
        scaling_factor = stage2_size / total_target
        target_counts = {k: int(v * scaling_factor) for k, v in target_counts.items()}

    # Sample according to targets
    stage2_frames = []
    for p_type, target in target_counts.items():
        if target > 0:
            type_df = remaining_df[remaining_df['product_type_clean'] == p_type]
            # Prioritize items with color when possible
            with_color = type_df[type_df['color_clean'] != 'MISSING']
            if len(with_color) >= target:
                stage2_frames.append(with_color.sample(target, random_state=42))
            else:
                # Take all with color, then fill in with items missing color
                stage2_frames.append(with_color)
                missing_color = type_df[type_df['color_clean'] == 'MISSING']
                needed = target - len(with_color)
                if len(missing_color) > 0 and needed > 0:
                    stage2_frames.append(missing_color.sample(min(len(missing_color), needed), random_state=42))

    stage2 = pd.concat(stage2_frames) if stage2_frames else pd.DataFrame()
    print(f"Stage 2 complete: {len(stage2)} items sampled")

    # STAGE 3: Rare categories and edge cases (10%)
    stage3_size = int(0.1 * n_samples)
    print(f"Stage 3: Finding rare categories and edge cases with {stage3_size} items")

    # Skip already sampled items
    remaining_df = clean_df[~clean_df.index.isin(pd.concat([stage1, stage2]).index)]

    # Find underrepresented categories
    common_types = clean_df['product_type_clean'].value_counts().head(50).index
    rare_df = remaining_df[~remaining_df['product_type_clean'].isin(common_types)]

    stage3 = rare_df.sample(min(len(rare_df), stage3_size), random_state=42)
    print(f"Stage 3 complete: {len(stage3)} items sampled")

    # Combine all stages
    final_sample = pd.concat([stage1, stage2, stage3])

    # Verify final distribution before returning
    final_product_counts = final_sample['product_type_clean'].value_counts()
    final_brand_counts = final_sample['brand_clean'].value_counts()
    final_color_counts = final_sample['color_clean'].value_counts()

    print("\nFinal distribution check:")
    print(f"SHOES: {final_product_counts.get('SHOES', 0)} (target: ~700)")
    print(f"AmazonBasics: {final_brand_counts.get('AmazonBasics', 0)} (target: ~1200)")
    print(f"Missing color: {final_color_counts.get('MISSING', 0)} (target: <3000)")

    # Optional: Make one final adjustment if severely off target
    if final_product_counts.get('SHOES', 0) > 800:
        excess = final_product_counts.get('SHOES', 0) - 700
        replace_shoes = final_sample[final_sample['product_type_clean'] == 'SHOES'].sample(
            min(excess, 100), random_state=42
        )

        # Find rare product types to replace with
        common_types = final_product_counts.head(10).index
        rare_remaining = clean_df[
            (~clean_df.index.isin(final_sample.index)) &
            (~clean_df['product_type_clean'].isin(common_types))
        ]

        if len(rare_remaining) > 0:
            replacements = rare_remaining.sample(min(len(rare_remaining), len(replace_shoes)), random_state=42)
            final_sample = final_sample[~final_sample.index.isin(replace_shoes.index)]
            final_sample = pd.concat([final_sample, replacements])
            print(f"Made final adjustment: Replaced {len(replacements)} excess SHOES with rare product types")


    # If we have too few, add random samples
    if len(final_sample) < n_samples:
        shortage = n_samples - len(final_sample)
        print(f"Need {shortage} more items to reach target {n_samples}")

        still_remaining = clean_df[~clean_df.index.isin(final_sample.index)]
        additional = still_remaining.sample(min(len(still_remaining), shortage), random_state=42)
        final_sample = pd.concat([final_sample, additional])

    # If we have too many, subsample
    if len(final_sample) > n_samples:
        print(f"Trimming {len(final_sample) - n_samples} excess items")
        final_sample = final_sample.sample(n_samples, random_state=42)

    print(f"Final sample size: {len(final_sample)}")

    # Return the IDs and cleaned DataFrame
    return final_sample[['item_id']].reset_index(drop=True), final_sample

# Execute the sampling
sample_ids, sample_df = balanced_multi_stage_sampling(df, n_samples=12000)

# Save the sampled IDs to use for image selection
sample_ids.to_csv('sampled_item_ids.csv', index=False)

# Optional: Generate summary of the sample
summary = {
    'product_type': sample_df['product_type_clean'].value_counts().head(10).to_dict(),
    'brand': sample_df['brand_clean'].value_counts().head(10).to_dict(),
    'color': sample_df['color_clean'].value_counts().head(10).to_dict()
}
print("\nSample Summary:")
for category, counts in summary.items():
    print(f"\nTop 10 {category}:")
    for value, count in counts.items():
        print(f"  {value}: {count}")
```
